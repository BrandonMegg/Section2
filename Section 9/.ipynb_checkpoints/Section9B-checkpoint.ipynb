```python
# reading MNIST Dataset 


import numpy as np # linear algebra
import struct
from array import array
from os.path  import join

#
# MNIST Data Loader Class
#
class MnistDataloader(object):
    def __init__(self, training_images_filepath,training_labels_filepath,
                 test_images_filepath, test_labels_filepath):
        self.training_images_filepath = training_images_filepath
        self.training_labels_filepath = training_labels_filepath
        self.test_images_filepath = test_images_filepath
        self.test_labels_filepath = test_labels_filepath
    
    def read_images_labels(self, images_filepath, labels_filepath):        
        labels = []
        with open(labels_filepath, 'rb') as file:
            magic, size = struct.unpack(">II", file.read(8))
            if magic != 2049:
                raise ValueError('Magic number mismatch, expected 2049, got {}'.format(magic))
            labels = array("B", file.read())        
        
        with open(images_filepath, 'rb') as file:
            magic, size, rows, cols = struct.unpack(">IIII", file.read(16))
            if magic != 2051:
                raise ValueError('Magic number mismatch, expected 2051, got {}'.format(magic))
            image_data = array("B", file.read())        
        images = []
        for i in range(size):
            images.append([0] * rows * cols)
        for i in range(size):
            img = np.array(image_data[i * rows * cols:(i + 1) * rows * cols])
            img = img.reshape(28, 28)
            images[i][:] = img            
        
        return images, labels
            
    def load_data(self):
        x_train, y_train = self.read_images_labels(self.training_images_filepath, self.training_labels_filepath)
        x_test, y_test = self.read_images_labels(self.test_images_filepath, self.test_labels_filepath)
        return (x_train, y_train),(x_test, y_test)
```

```python
import numpy as np
import struct
from array import array
from os.path import join
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay  


# MNIST Data Loader Class

class MnistDataloader(object):
    def __init__(self, training_images_filepath, training_labels_filepath,
                 test_images_filepath, test_labels_filepath):
        self.training_images_filepath = training_images_filepath
        self.training_labels_filepath = training_labels_filepath
        self.test_images_filepath = test_images_filepath
        self.test_labels_filepath = test_labels_filepath

    def read_images_labels(self, images_filepath, labels_filepath):        
        with open(labels_filepath, 'rb') as file:
            magic, size = struct.unpack(">II", file.read(8))
            if magic != 2049:
                raise ValueError(f'Magic number mismatch, expected 2049, got {magic}')
            labels = np.array(array("B", file.read()), dtype=np.int64)     

        with open(images_filepath, 'rb') as file:
            magic, size, rows, cols = struct.unpack(">IIII", file.read(16))
            if magic != 2051:
                raise ValueError(f'Magic number mismatch, expected 2051, got {magic}')
            image_data = np.array(array("B", file.read()), dtype=np.uint8).reshape(size, rows, cols)

        return image_data, labels

    def load_data(self):
        x_train, y_train = self.read_images_labels(self.training_images_filepath, self.training_labels_filepath)
        x_test, y_test = self.read_images_labels(self.test_images_filepath, self.test_labels_filepath)
        return (x_train, y_train), (x_test, y_test)




input_path = ''
training_images_filepath = join(input_path, 'train-images.idx3-ubyte')
training_labels_filepath = join(input_path, 'train-labels.idx1-ubyte')
test_images_filepath = join(input_path, 't10k-images.idx3-ubyte')
test_labels_filepath = join(input_path, 't10k-labels.idx1-ubyte')


#load and preprocess dataset
mnist_dataloader = MnistDataloader(training_images_filepath, training_labels_filepath, test_images_filepath, test_labels_filepath)
(x_train, y_train), (x_test, y_test) = mnist_dataloader.load_data()

#flatten images and normalize pixel values
x_train = x_train.reshape(-1, 28*28).astype(np.float32) / 255.0  
x_test = x_test.reshape(-1, 28*28).astype(np.float32) / 255.0

#split (80-20)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

print(f"x_train shape: {x_train.shape}, y_train shape: {y_train.shape}")
print(f"x_val shape: {x_val.shape}, y_val shape: {y_val.shape}")



class MLP:
    def __init__(self, input_size, hidden_sizes, output_size, activation='relu', learning_rate=0.01):
        self.activation = activation
        self.learning_rate = learning_rate
        
        
        layer_sizes = [input_size] + hidden_sizes + [output_size]
        
        #initialize weights and biases
        self.weights = [np.random.randn(layer_sizes[i], layer_sizes[i+1]) * 0.01 for i in range(len(layer_sizes) - 1)]
        self.biases = [np.zeros((1, layer_sizes[i+1])) for i in range(len(layer_sizes) - 1)]
        
    def relu(self, x):
        return np.maximum(0, x)
    
    def relu_derivative(self, x):
        return (x > 0).astype(float)
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_derivative(self, x):
        s = self.sigmoid(x)
        return s * (1 - s)
    
    def forward(self, X):
        activations = [X]
        pre_activations = []
        
        for W, b in zip(self.weights, self.biases):
            Z = np.dot(activations[-1], W) + b
            pre_activations.append(Z)
            
            if self.activation == 'relu':
                A = self.relu(Z)
            else:
                A = self.sigmoid(Z)
            
            activations.append(A)
        
        return activations, pre_activations
    
    def backward(self, X, y):
        activations, pre_activations = self.forward(X)
        
        # One-hot encode y
        y_one_hot = np.eye(self.weights[-1].shape[1])[y]
        
        dA = activations[-1] - y_one_hot
        dW, dB = [], []
        
        for i in reversed(range(len(self.weights))):
            Z = pre_activations[i]
            A_prev = activations[i]
            
            if self.activation == 'relu':
                dZ = dA * self.relu_derivative(Z)
            else:
                dZ = dA * self.sigmoid_derivative(Z)
            
            dW_i = np.dot(A_prev.T, dZ)
            dB_i = np.sum(dZ, axis=0, keepdims=True)
            
            dA = np.dot(dZ, self.weights[i].T)
            
            dW.insert(0, dW_i)
            dB.insert(0, dB_i)
        
        # Update weights and biases
        for i in range(len(self.weights)):
            self.weights[i] -= self.learning_rate * dW[i]
            self.biases[i] -= self.learning_rate * dB[i]
    
    def train(self, X, y, X_val, y_val, epochs=100):
        losses = []
        for epoch in range(epochs):
            self.backward(X, y)
            loss = self.compute_loss(X, y)
            losses.append(loss)
            if epoch % 10 == 0:
                print(f"Epoch {epoch}, Loss: {loss:.4f}")
        
        
        plt.figure()
        plt.plot(range(epochs), losses, label='Training Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Convergence as a function of Epoch')
        plt.legend()
        plt.savefig("convergence_plot.png")
        plt.show()
        
        # Compute confusion matrix
        y_pred = self.predict(X_val)
        cm = confusion_matrix(y_val, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        disp.plot()
        plt.savefig("confusion_matrix.png")
        plt.show()
    
    def compute_loss(self, X, y):
        activations, _ = self.forward(X)
        y_one_hot = np.eye(self.weights[-1].shape[1])[y]
        return np.mean(np.square(y_one_hot - activations[-1]))  
    
    def predict(self, X):
        activations, _ = self.forward(X)
        return np.argmax(activations[-1], axis=1)

# Train the MLP model
mlp = MLP(input_size=784, hidden_sizes=[128, 64], output_size=10, activation='relu', learning_rate=0.01)
mlp.train(x_train, y_train, x_val, y_val, epochs=100)
```

```python
import numpy as np
import struct
from array import array
from os.path import join
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

#MNIST Data Loader Class
class MnistDataloader:
    def __init__(self, training_images_filepath, training_labels_filepath,
                 test_images_filepath, test_labels_filepath):
        self.training_images_filepath = training_images_filepath
        self.training_labels_filepath = training_labels_filepath
        self.test_images_filepath = test_images_filepath
        self.test_labels_filepath = test_labels_filepath

    def read_images_labels(self, images_filepath, labels_filepath):        
        with open(labels_filepath, 'rb') as file:
            magic, size = struct.unpack(">II", file.read(8))
            if magic != 2049:
                raise ValueError(f'Magic number mismatch, expected 2049, got {magic}')
            labels = np.array(array("B", file.read()))

        with open(images_filepath, 'rb') as file:
            magic, size, rows, cols = struct.unpack(">IIII", file.read(16))
            if magic != 2051:
                raise ValueError(f'Magic number mismatch, expected 2051, got {magic}')
            images = np.array(array("B", file.read()), dtype=np.float32).reshape(size, rows * cols) / 255.0  # Normalize

        return images, labels

    def load_data(self):
        x_train, y_train = self.read_images_labels(self.training_images_filepath, self.training_labels_filepath)
        x_test, y_test = self.read_images_labels(self.test_images_filepath, self.test_labels_filepath)
        return (x_train, y_train), (x_test, y_test)

# Define File Paths
input_path = ''
training_images_filepath = join(input_path, 'train-images.idx3-ubyte')
training_labels_filepath = join(input_path, 'train-labels.idx1-ubyte')
test_images_filepath = join(input_path, 't10k-images.idx3-ubyte')
test_labels_filepath = join(input_path, 't10k-labels.idx1-ubyte')

# Load MNIST Data
mnist_dataloader = MnistDataloader(training_images_filepath, training_labels_filepath, test_images_filepath, test_labels_filepath)
(x_train, y_train), (x_test, y_test) = mnist_dataloader.load_data()

#split 80-20
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

# Define Multi-Layer Perceptron 
class MLP:
    def __init__(self, input_size, hidden_sizes, output_size, learning_rate=0.01, batch_size=32):
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        
        # Define layer sizes
        layer_sizes = [input_size] + hidden_sizes + [output_size]
        
        # Initialize weights and biases
        self.weights = [np.random.randn(layer_sizes[i], layer_sizes[i+1]) * np.sqrt(2.0 / layer_sizes[i]) 
                        for i in range(len(layer_sizes) - 1)]
        self.biases = [np.zeros((1, layer_sizes[i+1])) for i in range(len(layer_sizes) - 1)]

    def relu(self, x):
        return np.maximum(0, x)

    def relu_derivative(self, x):
        return (x > 0).astype(float)

    def softmax(self, x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def forward(self, X):
        activations = [X]
        pre_activations = []
        
        for W, b in zip(self.weights, self.biases):
            Z = np.dot(activations[-1], W) + b
            pre_activations.append(Z)
            A = self.relu(Z)
            activations.append(A)

        activations[-1] = self.softmax(pre_activations[-1])  #softmax at output layer
        return activations, pre_activations

    def compute_loss(self, X, y):
        activations, _ = self.forward(X)
        y_one_hot = np.eye(self.weights[-1].shape[1])[y]
        probs = activations[-1]
        loss = -np.mean(np.sum(y_one_hot * np.log(probs + 1e-8), axis=1))  
        return loss

    def backward(self, X, y):
        activations, pre_activations = self.forward(X)
        y_one_hot = np.eye(self.weights[-1].shape[1])[y]
        dA = activations[-1] - y_one_hot

        dW, dB = [], []

        for i in reversed(range(len(self.weights))):
            Z = pre_activations[i]
            A_prev = activations[i]

            dZ = dA * self.relu_derivative(Z) if i < len(self.weights) - 1 else dA
            dW_i = np.dot(A_prev.T, dZ)
            dB_i = np.sum(dZ, axis=0, keepdims=True)
            dA = np.dot(dZ, self.weights[i].T)

            dW.insert(0, dW_i)
            dB.insert(0, dB_i)

        for i in range(len(self.weights)):
            self.weights[i] -= self.learning_rate * dW[i]
            self.biases[i] -= self.learning_rate * dB[i]

    def train(self, X, y, X_valid, y_valid, epochs=100):
        losses = []

        for epoch in range(epochs):
            # mini batch training
            for i in range(0, X.shape[0], self.batch_size):
                X_batch = X[i:i + self.batch_size]
                y_batch = y[i:i + self.batch_size]
                self.backward(X_batch, y_batch)

            loss = self.compute_loss(X, y)
            losses.append(loss)

            if epoch % 10 == 0:
                print(f"Epoch {epoch}, Loss: {loss:.4f}")

        
        plt.figure()
        plt.plot(range(epochs), losses, label='Training Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Convergence as a function of Epoch')
        plt.legend()
        plt.savefig("convergence_plot.png")
        plt.show()

        #compute Confusion Matrix
        y_pred = self.predict(X_valid)
        cm = confusion_matrix(y_valid, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        disp.plot()
        plt.savefig("confusion_matrix_MLP.png")
        plt.show()

    def predict(self, X):
        activations, _ = self.forward(X)
        return np.argmax(activations[-1], axis=1)

#Initialize and Train MLP
mlp = MLP(input_size=784, hidden_sizes=[256, 128, 64], output_size=10, learning_rate=0.01, batch_size=32)
mlp.train(x_train, y_train, x_valid, y_valid, epochs=100)
```

```python
#task 2 and 3 

import numpy as np
import struct
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay


class MnistDataloader:
    def __init__(self, img_train_path, lbl_train_path, img_test_path, lbl_test_path):
        self.img_train_path = img_train_path
        self.lbl_train_path = lbl_train_path
        self.img_test_path = img_test_path
        self.lbl_test_path = lbl_test_path

    def read_images_labels(self, img_path, lbl_path):
        with open(lbl_path, 'rb') as f:
            _, size = struct.unpack(">II", f.read(8))
            labels = np.array(list(f.read(size)))

        with open(img_path, 'rb') as f:
            _, size, rows, cols = struct.unpack(">IIII", f.read(16))
            images = np.array(list(f.read(size * rows * cols))).reshape(size, rows, cols) / 255.0  # Normalize

        return images, labels

    def load_data(self):
        x_train, y_train = self.read_images_labels(self.img_train_path, self.lbl_train_path)
        x_test, y_test = self.read_images_labels(self.img_test_path, self.lbl_test_path)
        return (x_train, y_train), (x_test, y_test)

#convolution function
def conv2d(X, W, stride=1):
    h, w = W.shape
    X_h, X_w = X.shape
    out_h = (X_h - h) // stride + 1
    out_w = (X_w - w) // stride + 1
    out = np.zeros((out_h, out_w))

    for i in range(0, out_h):
        for j in range(0, out_w):
            out[i, j] = np.sum(X[i * stride:i * stride + h, j * stride:j * stride + w] * W)
    return out

def maxpool2d(X, size=2, stride=2):
    X_h, X_w = X.shape
    out_h = (X_h - size) // stride + 1
    out_w = (X_w - size) // stride + 1
    out = np.zeros((out_h, out_w))

    for i in range(0, out_h):
        for j in range(0, out_w):
            out[i, j] = np.max(X[i * stride:i * stride + size, j * stride:j * stride + size])
    return out

#activation functions
def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exps = np.exp(x - np.max(x))
    return exps / np.sum(exps)


class CNN:
    def __init__(self):
        self.conv1 = np.random.randn(3, 3) * 0.1
        self.fc1 = np.random.randn(13 * 13, 128) * 0.1
        self.fc2 = np.random.randn(128, 10) * 0.1

    def forward(self, X):
        self.X = X
        self.conv_out = relu(conv2d(X, self.conv1))
        self.pool_out = maxpool2d(self.conv_out)
        self.fc_input = self.pool_out.flatten()
        self.fc1_out = relu(np.dot(self.fc_input, self.fc1))
        self.fc2_out = np.dot(self.fc1_out, self.fc2)
        self.output = softmax(self.fc2_out)
        return self.output

    def backward(self, y_true, lr=0.01):
        y_one_hot = np.zeros(10)
        y_one_hot[y_true] = 1
        d_fc2 = self.output - y_one_hot
        d_fc1 = np.dot(d_fc2, self.fc2.T) * (self.fc1_out > 0)
        d_fc_input = np.dot(d_fc1, self.fc1.T).reshape(self.pool_out.shape)
        
        d_pool = np.zeros_like(self.pool_out)
        for i in range(d_pool.shape[0]):
            for j in range(d_pool.shape[1]):
                d_pool[i, j] = d_fc_input[i, j] * (self.pool_out[i, j] == self.conv_out[i*2:i*2+2, j*2:j*2+2].max())
        
        d_conv = np.zeros_like(self.conv1)
        for i in range(d_conv.shape[0]):
            for j in range(d_conv.shape[1]):
                d_conv[i, j] = np.sum(d_pool * self.X[i:i+13, j:j+13])
        
        self.fc2 -= lr * np.outer(self.fc1_out, d_fc2)
        self.fc1 -= lr * np.outer(self.fc_input, d_fc1)
        self.conv1 -= lr * d_conv

    def train(self, X_train, y_train, X_test, y_test, epochs=10, lr=0.01):
        losses = []
        for epoch in range(epochs):
            loss = 0
            for i in range(len(X_train)):
                output = self.forward(X_train[i])
                loss += -np.log(output[y_train[i]] + 1e-9)
                self.backward(y_train[i], lr)
            loss /= len(X_train)
            losses.append(loss)
            print(f"Epoch {epoch}, Loss: {loss:.4f}")

        # Plot Loss Curve
        plt.figure()
        plt.plot(range(epochs), losses, label='Training Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Convergence as a Function of Epoch')
        plt.legend()
        plt.savefig("cnn_convergence_plot.png")
        plt.show()

        # Confusion Matrix
        y_pred = np.array([np.argmax(self.forward(X)) for X in X_test])
        cm = confusion_matrix(y_test, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        disp.plot()
        plt.savefig("cnn_confusion_matrix.png")
        plt.show()

# Load MNIST Data
input_path = ''
train_img = input_path + 'train-images.idx3-ubyte'
train_lbl = input_path + 'train-labels.idx1-ubyte'
test_img = input_path + 't10k-images.idx3-ubyte'
test_lbl = input_path + 't10k-labels.idx1-ubyte'

mnist_loader = MnistDataloader(train_img, train_lbl, test_img, test_lbl)
(X_train, y_train), (X_test, y_test) = mnist_loader.load_data()

#flatten images for CNN
X_train = np.array([x / 255.0 for x in X_train])
X_test = np.array([x / 255.0 for x in X_test])

#Train CNN Model
cnn = CNN()
cnn.train(X_train[:1000], y_train[:1000], X_test[:200], y_test[:200], epochs=10, lr=0.01)
```
