```python
#a
#thermal convection 
#x is rate of convection motion 
#y is the temp difference between currents 
#z difference from thermal equilibrium 
#σ ratio of viscosity to thermal diffusivity 
#β describes the effects of height to width ratio of a convective cell on the heat transfer between cells
```

```python
#b 

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def lorenz(t, state, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]


sigma = 10
rho = 48
beta = 3


initial_state = [1.0, 1.0, 1.0]

t_span = (0, 12)
t_eval = np.linspace(*t_span, 5000)  

solution = solve_ivp(lorenz, t_span, initial_state, t_eval=t_eval, args=(sigma, rho, beta))

x, y, z = solution.y

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5, color='blue')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Lorenz Attractor (σ=10, ρ=48, β=3)")
plt.show()
plt.savefig("Lorentz Attractor")

lorenz_attractor_coords = np.column_stack((x, y, z))
np.savetxt("lorenz_attractor.csv", lorenz_attractor_coords, delimiter=",", header="x,y,z", comments="")
```

```python
#c 

import matplotlib.animation as animation

def lorenz(t, state, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]


sigma = 10
rho = 48
beta = 3

initial_state = [1.0, 1.0, 1.0]

t_span = (0, 12)
t_eval = np.linspace(*t_span, 2500)  


solution = solve_ivp(lorenz, t_span, initial_state, t_eval=t_eval, args=(sigma, rho, beta))

x, y, z = solution.y


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Lorenz Attractor Time Evolution (σ=10, ρ=48, β=3)")


line, = ax.plot([], [], [], lw=1.5, color='blue')

#animation function
def update(num):
    line.set_data(x[:num], y[:num])
    line.set_3d_properties(z[:num])
    return line,


ani = animation.FuncAnimation(fig, update, frames=len(t_eval), interval=10, blit=True)


ani.save("lorenz_attractor.mp4", writer="ffmpeg", fps=20)

plt.show()

lorenz_attractor_coords = np.column_stack((x, y, z))
np.savetxt("lorenz_attractor.csv", lorenz_attractor_coords, delimiter=",", header="x,y,z", comments="")
```
