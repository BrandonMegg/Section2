```python
#statistical distribution of the Ising model

import numpy as np
import matplotlib.pyplot as plt

#parameters
L = 4  #lattice size
B = 0  #external magnetic field
J = 1  #interaction strength
T = 1  #tempe
beta = 1 / T  #inverse temperature

#creates the lattice of spins 
def initialize_spins(L):
    return np.random.choice([-1, 1], size=(L, L))

# Hamiltonian function for Ising model
def hamiltonian(spins, J=1, B=0):
    H = 0
    #sums over nearest neighbors
    for i in range(L):
        for j in range(L):
            neighbors = [(i+1)%L, j], [(i-1)%L, j], [i, (j+1)%L], [i, (j-1)%L]
            for ni, nj in neighbors:
                H -= J * spins[i, j] * spins[ni, nj]
            H -= B * spins[i, j]
    return H

#Metropolis Hastings algorithm to sample from the distribution
def metropolis_hastings(L, beta, num_samples):
    spins = initialize_spins(L)
    samples = []
    for _ in range(num_samples):
        #randomly select a spin to flip
        i, j = np.random.randint(0, L, 2)
        new_spins = spins.copy()
        new_spins[i, j] = -spins[i, j]  #flips the selected spin
        
        #calculate energy change
        delta_H = hamiltonian(new_spins) - hamiltonian(spins)
        
        #accept or reject the flip based on Metropolis 
        if delta_H < 0 or np.random.rand() < np.exp(-beta * delta_H):
            spins = new_spins  #change to new configuration
            
        samples.append(spins.copy())
    
    return np.array(samples)

#sampling spin configurations
num_samples = 10000  
samples = metropolis_hastings(L, beta, num_samples)

#plotting the samples
fig, ax = plt.subplots(figsize=(8, 6))
for sample in samples[::500]:  #take every 500th sample for visualization
    ax.imshow(sample, cmap='coolwarm', interpolation='nearest')
ax.set_title("Samples of Spin Configurations")
ax.axis('off')
plt.savefig("ising_samples.png")

#plotting the histogram of energy values
energies = [hamiltonian(s) for s in samples]
plt.figure(figsize=(8, 6))
plt.hist(energies, bins=50, density=True, alpha=0.6, color='g')
plt.xlabel('Energy')
plt.ylabel('Probability Density')
plt.title("Energy Distribution of Ising Model")
plt.savefig("energy_distribution.png")

plt.show()
```

```python
#Gibbs sampler on Ising model and Gibbs iteration


L = 4  # Lattice size
J = 1  # Interaction strength
T = 1  # Temperature
beta = 1 / T  # Inverse temperature
num_samples = 10000  # Number of samples to generate
burn_in_periods = [1000, 3000, 5000]  # Different burn-in periods
num_chains = 4  # Number of parallel chains

# Function to initialize spins
def initialize_spins(L):
    return np.random.choice([-1, 1], size=(L, L))

# Hamiltonian function for Ising model (without B, since B=0)
def hamiltonian(spins, J=1):
    H = 0
    for i in range(L):
        for j in range(L):
            neighbors = [(i+1)%L, j], [(i-1)%L, j], [i, (j+1)%L], [i, (j-1)%L]
            for ni, nj in neighbors:
                H -= J * spins[i, j] * spins[ni, nj]
    return H

# Function to calculate the conditional probability for a given spin
def conditional_prob(spins, i, j, beta, J=1):
    neighbors = [(i+1)%L, j], [(i-1)%L, j], [i, (j+1)%L], [i, (j-1)%L]
    energy_up = 2 * J * spins[i, j] * sum([spins[ni, nj] for ni, nj in neighbors])
    p_up = np.exp(beta * energy_up) / (np.exp(beta * energy_up) + np.exp(-beta * energy_up))
    return p_up

# Gibbs sampling for multiple chains
def gibbs_sampling(num_samples, burn_in_period, num_chains, L, beta):
    chains = [initialize_spins(L) for _ in range(num_chains)]
    all_samples = []
    
    for _ in range(num_samples):
        for chain in chains:
            # Update each spin sequentially
            for i in range(L):
                for j in range(L):
                    p_up = conditional_prob(chain, i, j, beta)
                    # Sample new value for spin at (i,j)
                    chain[i, j] = 1 if np.random.rand() < p_up else -1
            # Collect the chain configuration after burn-in period
            all_samples.append(chain.copy())
    
    return np.array(all_samples)

# Sampling
all_samples = gibbs_sampling(num_samples, max(burn_in_periods), num_chains, L, beta)

# Plotting: Visualization of spin configurations with different burn-in periods
fig, axs = plt.subplots(len(burn_in_periods), num_chains, figsize=(12, 6))
for idx, burn_in in enumerate(burn_in_periods):
    # Discarding burn-in samples
    samples_after_burn_in = all_samples[burn_in:]
    for chain_idx, chain in enumerate(samples_after_burn_in[::100]):  # take every 100th sample
        axs[idx, chain_idx].imshow(chain, cmap='coolwarm', interpolation='nearest')
        axs[idx, chain_idx].set_title(f"Chain {chain_idx+1}")
        axs[idx, chain_idx].axis('off')
    axs[idx, 0].set_ylabel(f"Burn-in {burn_in}")

plt.tight_layout()
plt.savefig("ising_sampling_burn_in.png")
plt.show()

# Plotting Energy Distribution
energies = np.array([hamiltonian(chain) for chain in all_samples])
plt.figure(figsize=(8, 6))
plt.hist(energies, bins=50, density=True, alpha=0.6, color='g')
plt.xlabel('Energy')
plt.ylabel('Probability Density')
plt.title("Energy Distribution of Ising Model")
plt.savefig("energy_distribution_burn_in.png")
plt.show()
```

```python
#Magnetization, Landau Theory, and Phase Transition of the Ising Model 

L_values = [10, 17, 25, 32, 40]  
T_values = np.linspace(0.5, 5.0, 20)  
J = 1  
beta_values = 1 / T_values  
num_samples = 5000  
burn_in = 2000  


def initialize_spins(L):
    return np.random.choice([-1, 1], size=(L, L))


def conditional_prob(spins, i, j, beta, J=1):
    L = spins.shape[0]
    neighbors = [(i+1)%L, j], [(i-1)%L, j], [i, (j+1)%L], [i, (j-1)%L]
    energy_up = 2 * J * spins[i, j] * sum([spins[ni, nj] for ni, nj in neighbors])
    p_up = np.exp(beta * energy_up) / (np.exp(beta * energy_up) + np.exp(-beta * energy_up))
    return p_up


def gibbs_sampling(L, beta, num_samples, burn_in):
    spins = initialize_spins(L)
    magnetization = []
    
    for _ in range(num_samples):
        for i in range(L):
            for j in range(L):
                p_up = conditional_prob(spins, i, j, beta)
                spins[i, j] = 1 if np.random.rand() < p_up else -1
        
        if _ > burn_in:  
            magnetization.append(np.abs(np.sum(spins) / (L * L)))
    
    return np.mean(magnetization), spins

#run the simulations
results = {}
for L in L_values:
    M_values = []
    for beta in beta_values:
        M, final_spins = gibbs_sampling(L, beta, num_samples, burn_in)
        M_values.append(M)
    results[L] = M_values
    

plt.figure(figsize=(8, 6))
for L in L_values:
    plt.plot(T_values, results[L], label=f'L={L}')
plt.xlabel('Temperature T')
plt.ylabel('Magnetization M')
plt.title('Ferromagnetic to Paramagnetic Phase Transition')
plt.legend()
plt.grid()
plt.savefig("ising_phase_transition.png")
plt.show()
```

```python
#Classical Magnetic Field dependence: magnetization of the 2D Ising model 


L_values = [10, 17, 25, 32, 40]  
T = 2.5  
J = 1  
B_values = np.linspace(-2, 2, 20)  
beta = 1 / T  
num_samples = 5000  
burn_in = 2000  


def initialize_spins(L):
    return np.random.choice([-1, 1], size=(L, L))

#conditional probability with external field B
def conditional_prob(spins, i, j, beta, B, J=1):
    L = spins.shape[0]
    neighbors = [(i+1)%L, j], [(i-1)%L, j], [i, (j+1)%L], [i, (j-1)%L]
    energy_up = 2 * J * spins[i, j] * sum([spins[ni, nj] for ni, nj in neighbors]) + 2 * B * spins[i, j]
    p_up = np.exp(beta * energy_up) / (np.exp(beta * energy_up) + np.exp(-beta * energy_up))
    return p_up

def gibbs_sampling(L, beta, B, num_samples, burn_in):
    spins = initialize_spins(L)
    magnetization = []
    
    for _ in range(num_samples):
        for i in range(L):
            for j in range(L):
                p_up = conditional_prob(spins, i, j, beta, B)
                spins[i, j] = 1 if np.random.rand() < p_up else -1
        
        if _ > burn_in:  
            magnetization.append(np.abs(np.sum(spins) / (L * L)))
    
    return np.mean(magnetization)


results = {}
for L in L_values:
    M_values = []
    for B in B_values:
        M = gibbs_sampling(L, beta, B, num_samples, burn_in)
        M_values.append(M)
    results[L] = M_values
    

plt.figure(figsize=(8, 6))
for L in L_values:
    plt.plot(B_values, results[L], label=f'L={L}')
plt.xlabel('Magnetic Field B')
plt.ylabel('Magnetization M')
plt.title('Magnetization vs Magnetic Field at T=2.5')
plt.legend()
plt.grid()
plt.savefig("ising_m_vs_b.png")
plt.show()
```
